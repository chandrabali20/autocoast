```python
def crop_n_zoom(imgfile,a,b,c,d):
    from PIL import Image
    img = Image.open(imgfile)
    area = (a, b, c, d)
    cropped_img = img.crop(area)
    cropped_n_resized_img = cropped_img.resize((400, 400), Image.LANCZOS)
    return cropped_n_resized_img
```
